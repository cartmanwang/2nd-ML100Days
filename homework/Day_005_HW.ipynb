```python
# 作業目標 1
print("Mean of 'CNT_CHILDREN' =", app_train['CNT_CHILDREN'].mean())
print("Std of 'CNT_CHILDREN' =", app_train['CNT_CHILDREN'].std())
```

```python
# 作業目標 2
app_train['CNT_CHILDREN'].hist()
```

```python
fig, axis = plt.subplots()

axis.yaxis.grid(True)
axis.set_title('CNT_CHILDREN',fontsize=10)
axis.set_xlabel("Count of Children",fontsize=10)
axis.set_ylabel("#Counts",fontsize=10)
#axis.set_xlim(0, app_train['CNT_CHILDREN'].max())

plt.hist(app_train['CNT_CHILDREN'], bins=19, log=True, facecolor='green')
```

```python
# 看看平均每個申請者擁有的小孩數目是否跟學歷有關連性, 看起來有相關
print("Mean of CNT_CHILDREN =", app_train['CNT_CHILDREN'].mean())
print()
print(app_train.groupby(by='NAME_EDUCATION_TYPE')['CNT_CHILDREN'].mean())
```

```python
app_train.groupby(by='NAME_EDUCATION_TYPE')['CNT_CHILDREN'].mean().plot(kind='bar')
```
